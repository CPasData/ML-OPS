#### <font color= #e1efa9>**Pruebas en Local**</font><br>

```python
import requests

# Ruta 4 - POST JSON

url = "http://127.0.0.1:5000/api/v1/predict"

datos = {
    "credit_score": 721,
    "country": "Spain",
    "gender": "Male",
    "age": 75,
    "tenure": 7,
    "balance": 64000,
    "products_number": 1,
    "credit_card": 1,
    "active_member": 1,
    "estimated_salary": 42500
}

respuesta = requests.post(url, json=datos)
print("Status code:", respuesta.status_code)
print("Respuesta:", respuesta.json())
```

```python
# Ruta 5 - Historial de predicciones

respuesta = requests.get("http://127.0.0.1:5000/api/v1/predicciones")
print("Status code:", respuesta.status_code)
print("Respuesta:", respuesta.json())
```

```python
# Ruta 6 - Contador de predicciones

respuesta = requests.get("http://127.0.0.1:5000/api/v1/predicciones/count")
print("Status code:", respuesta.status_code)
print("Respuesta:", respuesta.json())
```
